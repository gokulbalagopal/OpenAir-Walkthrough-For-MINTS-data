```r
#Loading the libraries
library(openair)
library(dplyr)
library(openairmaps)
library(latex2exp)
library(worldmet)
library(data.table)
```

```r
data_frame = read.csv("D:\\UTD\\UTDFall2022\\VariogramsLoRa\\firmware\\data\\Parameters\\csv\\Joappa_2022.csv")
df = as_tibble(data_frame)
df$date = as.POSIXct(df$dateTime,format="%Y-%m-%dT%H:%M:%S")
df = subset(df, select = -c(dateTime))
df_min <- timeAverage(df, avg.time = "min")
colnames(df_min)= c("date","pm0.1","pm0.3","pm1.0","pm2.5","pm10")
df_min
```

```r
#Loading the libraries
library(openair)
library(dplyr)
library(openairmaps)
library(latex2exp)
library(worldmet)
library(data.table)
```

```r
data_frame = read.csv("D:\\UTD\\UTDFall2022\\VariogramsLoRa\\firmware\\data\\Parameters\\csv\\Joappa_2022.csv")
```

```r
data_frame
```

```r
df = as_tibble(data_frame)
df$date = as.POSIXct(df$dateTime,format="%Y-%m-%dT%H:%M:%S")
df = subset(df, select = -c(dateTime))
```

```r
df_min <- timeAverage(df, avg.time = "min")
colnames(df_min)= c("date","pm0.1","pm0.3","pm1.0","pm2.5","pm10")
df_min
```

# Time Plots

```r
timePlot(df_min, 
         pollutant = c("pm2.5", "pm10"), 
         y.relation = "free")
```

```r
timePlot(selectByDate(df_min, year = 2022, month = "dec"),
         pollutant = c("pm2.5", "pm10"),
         y.relation = "free")
```

# Temporal Variation Plots

```r
#The shading shows the 95% confidence intervals of the mean
timeVariation(df_min, 
              pollutant = c("pm0.1", "pm0.3", "pm1.0", "pm2.5"), 
              normalise = TRUE)
```

```r
df_sp <- splitByDate(df_min, dates= "15/12/2022",
                        labels = c("before 15th Dec. 2022", "After 15th Dec. 2022"))
```

```r
df_sp
```

```r
timeVariation(df_sp, pollutant = "pm2.5", 
              group = "split.by", 
              difference = TRUE)
```

```r
timeVariation(df_min, pollutant = "pm10", 
              statistic = "median",
              col = "firebrick")
```

```r
myOutput <- timeVariation(df_min, pollutant = "pm2.5")

head(myOutput$data$day.hour)
```

```r
## just the diurnal variation
plot(myOutput, subset = "hour")
```

```r
## day and hour
plot(myOutput, subset = "day.hour")
```

```r
## weekday variation
plot(myOutput, subset = "day")
```

```r
## monthly variation
plot(myOutput, subset = "month")
```
